```python
# create X and y
X=df.drop("Chance of Admit ",axis=1)
y=df["Chance of Admit "]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=None)

# modelling
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

print("Intercept:",model.intercept_)
print("Coefficeants:",model.coef_)

# evaluation
# predictions
ypred_test = model.predict(X_test)

# mean absolute error
from sklearn.metrics import mean_absolute_error
print("Test MAE:",mean_absolute_error(y_test,ypred_test))

from sklearn.metrics import mean_squared_error
print("Test MSE:",mean_squared_error(y_test,ypred_test))

print("Test RMSE:",np.sqrt(mean_squared_error(y_test,ypred_test)))

# find test r square =R^2
print("Test R2:",model.score(X_test,y_test))

```

```python
# model selection
# 1. ckecklist first : model is good or aither overfitting or under fitting problem
# test accuracy = train accuracy

# prediction on train data
ypred_train = model.predict(X_train)

# train R2
print("Train R2:",model.score(X_train,y_train))
```

```python
# checklist 2: Test Accuracy == Cross Validation Score

from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(model,X,y,cv=5)
print(cv_score)

print("Cross_Val_Scrore:",cv_score.mean())


```

```python
# Model 3 :- Polynomial regression
# X & y
X=df.drop("Chance of Admit ",axis=1)
y=df["Chance of Admit "]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)

# Modelling
# Data preprocessing on train data

from sklearn.preprocessing import PolynomialFeatures
polynomial_converter=PolynomialFeatures(degree=2)
X_train_poly=pd.DataFrame(polynomial_converter.fit_transform(X_train))

# Modelling on Train Data

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train_poly,y_train)

# prediction on train data
ypred_train=model.predict(X_train_poly)

# evaluation on train data
from sklearn.metrics import mean_squared_error,r2_score
print("Train RMSE",np.sqrt(mean_squared_error(y_train,ypred_train)))
print("Test R2",r2_score(y_train,ypred_train))

# cross validation on train data

from sklearn.model_selection import cross_val_score
print("Cross_val_Score",cross_val_score(model,X_train_poly,y_train,cv=5).mean())


# data preprocessing on test data
X_test_poly = pd.DataFrame(polynomial_converter.transform(X_test))

# prediction on test data
ypred_test = model.predict(X_test_poly)

# evaluation on test data
print("Test RMSE",np.sqrt(mean_squared_error(y_test,ypred_test)))
print("Test R2",r2_score(y_test,ypred_test))

# hyperparameter tuning
# choosing the best polynomial degree for given data

train_r2=[]
test_r2=[]

for i in range(1,10):
  # preprocessing
  polynomial_converter = PolynomialFeatures(degree=i)
  X_train_poly = pd.DataFrame(polynomial_converter.fit_transform(X_train))

  # modelling

  model = LinearRegression()
  model.fit(X_train_poly,y_train)

  # prediction and evaluation on train data

  train_pred = model.predict(X_train_poly)
  train_r2.append(model.score(X_train_poly,y_train))

  # transformation on test data
  X_test_poly = pd.DataFrame(polynomial_converter.transform(X_test))

  # prediction and evalution on train data
  test_pred = model.predict(X_test_poly)
  test_r2.append(model.score(X_test_poly,y_test))

train_r2

test_r2
```
